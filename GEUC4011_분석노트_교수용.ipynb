## Data analysis with one variable

```python
def count (df, number) :
    return df.iloc[:,number].value_counts()
```

```python
def count_plot(df, number):
    # Get the column values
    column_values = df.iloc[:, number].value_counts().sort_index()
    
    # Generate the plot
    plt.figure(figsize=(10,6))
    sns.barplot(x=column_values.index, y=column_values.values)
    
    # Set the title
    plt.title(df.columns[number])
    
    # Show the plot
    plt.show()
```
